```python
import random
from pathlib import Path
from ultralytics import YOLO
from IPython.display import display, Image
```

```python
model_path = r"C:\Users\misha\OneDrive\Desktop\DL\yolo-damage\runs\damage_yolo_run1-3\weights\best.pt"
val_images_dir = Path(r"C:\Users\misha\OneDrive\Desktop\DL\yolo-damage\data\val\images")
png_files = list(val_images_dir.glob("*.png"))

if not png_files:
    print(f"Error, no images found in {val_images_dir}")
else:
    image_path = random.choice(png_files)
    print(f"Chosen image: {image_path.name}")

    model = YOLO(model_path)
    
    results = model.predict(source=str(image_path), save=True, imgsz=1024, conf=0.25)

    save_dir = Path(results[0].save_dir)
    
    saved_files = list(save_dir.glob("*.*"))
    
    if saved_files:
        result_img_path = saved_files[0]
        print(f"\nFile successfully found: {result_img_path.name}")
        display(Image(filename=str(result_img_path)))
    else:
        print(f"No files found in: {save_dir}")
```
